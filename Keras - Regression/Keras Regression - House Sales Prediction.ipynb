# The Data

 In this project of House Sales we should predict the price of the houses based on the features shown below. 
 
 We will be using data from a Kaggle data set.

**Feature Columns**

* id - Unique ID for each home sold
* date - Date of the home sale
* price - Price of each home sold
* bedrooms - Number of bedrooms
* bathrooms - Number of bathrooms, where .5 accounts for a room with a toilet but no shower
* sqft_living - Square footage of the apartments interior living space
* sqft_lot - Square footage of the land space
* floors - Number of floors
* waterfront - A dummy variable for whether the apartment was overlooking the waterfront or not
* view - An index from 0 to 4 of how good the view of the property was
* condition - An index from 1 to 5 on the condition of the apartment,
* grade - An index from 1 to 13, where 1-3 falls short of building construction and design, 7 has an average level of construction and design, and 11-13 have a high quality level of construction and design.
* sqft_above - The square footage of the interior housing space that is above ground level
* sqft_basement - The square footage of the interior housing space that is below ground level
* yr_built - The year the house was initially built
* yr_renovated - The year of the house’s last renovation
* zipcode - What zipcode area the house is in
* lat - Lattitude
* long - Longitude
* sqft_living15 - The square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15 - The square footage of the land lots of the nearest 15 neighbors

# Working with Feature Data

```python
df = df.drop('id', axis=1)
```

```python
df = df.drop('zipcode', axis=1)
```

```python
df.head(5)
```

## Feature Engineering from Date

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].apply(lambda date:date.month)
```

```python
df['year'] = df['date'].apply(lambda date:date.year)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
```

This plot definitely makes sense because if you look back at King County sales they're just increasing in price as time goes by.

```python
df = df.drop('date', axis=1)
```

```python
df.head(5)
```

# Train Test Split

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

# Creating a Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

# Training the Model

```python
model.fit(x=X_train, y=y_train.values, 
          validation_data=(X_test, y_test.values),
         batch_size=128, epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

# Evaluating on Test Data

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

## Predicting on Brand New Data

```python
X_test
```

```python
predictions = model.predict(X_test)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
explained_variance_score(y_test,predictions)
```

```python
df['price'].mean()
```

```python
df['price'].median()
```

```python
# Our predictions
plt.scatter(y_test, predictions)

# Perfect predictions
plt.plot(y_test, y_test, 'r')
```

So at these really expensive houses, we're actually not good at predicting the price, but we're pretty good at predicting the prices of houses between zero and $2 million.

```python
errors = y_test.values.reshape(6480, 1) - predictions
```

```python
sns.distplot(errors, kde=True, bins=50 )
```
